### Importación de datos



```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación



En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
total_tienda1 = tienda1['Precio'].sum()
total_tienda2 = tienda2['Precio'].sum()
total_tienda3 = tienda3['Precio'].sum()
total_tienda4 = tienda4['Precio'].sum()
```

```python
print(f' La suma total de la tienda1 es: ${total_tienda1:,.2f}')
print(f' La suma total de la tienda2 es: ${total_tienda2:,.2f}')
print(f' La suma total de la tienda3 es: ${total_tienda3:,.2f}')
print(f' La suma total de la tienda4 es: ${total_tienda4:,.2f}')
```

```python
def format_latam(number):
    # Format with default (English) locale: comma for thousands, period for decimal
    formatted_str = f"{number:,.2f}"
    # Replace thousands separator (comma) with a temporary character
    formatted_str = formatted_str.replace(",", "#")
    # Replace decimal separator (period) with a comma
    formatted_str = formatted_str.replace(".", ",")
    # Replace temporary character with a period
    formatted_str = formatted_str.replace("#", ".")
    return formatted_str
```

```python
print(f' La suma total de la tienda1 es: ${format_latam(total_tienda1)}')
print(f' La suma total de la tienda2 es: ${format_latam(total_tienda2)}')
print(f' La suma total de la tienda3 es: ${format_latam(total_tienda3)}')
print(f' La suma total de la tienda4 es: ${format_latam(total_tienda4)}')
```

# 2. Ventas por categoría

En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda

```python
ventas_por_categoria_tienda1 = tienda1.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
ventas_por_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
ventas_por_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
ventas_por_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)

print("Ventas por Categoría - Tienda 1:")
print(ventas_por_categoria_tienda1.apply(format_latam).to_string()) #Coloco el to_string para eliminar los metadatod de Name: Precio y dtype
print("\nVentas por Categoría - Tienda 2:")
print(ventas_por_categoria_tienda2.apply(format_latam).to_string())
print("\nVentas por Categoría - Tienda 3:")
print(ventas_por_categoria_tienda3.apply(format_latam).to_string())
print("\nVentas por Categoría - Tienda 4:")
print(ventas_por_categoria_tienda4.apply(format_latam).to_string())
```

### Conteo de productos vendidos por categoría

```python
conteo_productos_tienda1 = tienda1.groupby('Categoría del Producto').size().sort_values(ascending=False)
conteo_productos_tienda2 = tienda2.groupby('Categoría del Producto').size().sort_values(ascending=False)
conteo_productos_tienda3 = tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False)
conteo_productos_tienda4 = tienda4.groupby('Categoría del Producto').size().sort_values(ascending=False)

print("Conteo de Productos por Categoría - Tienda 1:")
print(conteo_productos_tienda1.to_string())
print("\nConteo de Productos por Categoría - Tienda 2:")
print(conteo_productos_tienda2.to_string())
print("\nConteo de Productos por Categoría - Tienda 3:")
print(conteo_productos_tienda3.to_string())
print("\nConteo de Productos por Categoría - Tienda 4:")
print(conteo_productos_tienda4.to_string())
```

# 3. Calificación promedio de la tienda


En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
calificacion_promedio_tienda1 = tienda1['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()

print(f'Calificación promedio de la Tienda 1: {calificacion_promedio_tienda1:.2f}')
print(f'Calificación promedio de la Tienda 2: {calificacion_promedio_tienda2:.2f}')
print(f'Calificación promedio de la Tienda 3: {calificacion_promedio_tienda3:.2f}')
print(f'Calificación promedio de la Tienda 4: {calificacion_promedio_tienda4:.2f}')
```

# 4. Productos más y menos vendidos

En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
conteo_productos_tienda1_df = tienda1.groupby('Producto').size().sort_values(ascending=False)
top_5_tienda1 = conteo_productos_tienda1_df.head(5)
bottom_5_tienda1 = conteo_productos_tienda1_df.tail(5)

conteo_productos_tienda2_df = tienda2.groupby('Producto').size().sort_values(ascending=False)
top_5_tienda2 = conteo_productos_tienda2_df.head(5)
bottom_5_tienda2 = conteo_productos_tienda2_df.tail(5)

conteo_productos_tienda3_df = tienda3.groupby('Producto').size().sort_values(ascending=False)
top_5_tienda3 = conteo_productos_tienda3_df.head(5)
bottom_5_tienda3 = conteo_productos_tienda3_df.tail(5)

conteo_productos_tienda4_df = tienda4.groupby('Producto').size().sort_values(ascending=False)
top_5_tienda4 = conteo_productos_tienda4_df.head(5)
bottom_5_tienda4 = conteo_productos_tienda4_df.tail(5)

print("Top 5 productos más vendidos - Tienda 1:")
print(top_5_tienda1)
print("\nBottom 5 productos menos vendidos - Tienda 1:")
print(bottom_5_tienda1)

print("\nTop 5 productos más vendidos - Tienda 2:")
print(top_5_tienda2)
print("\nBottom 5 productos menos vendidos - Tienda 2:")
print(bottom_5_tienda2)

print("\nTop 5 productos más vendidos - Tienda 3:")
print(top_5_tienda3)
print("\nBottom 5 productos menos vendidos - Tienda 3:")
print(bottom_5_tienda3)

print("\nTop 5 productos más vendidos - Tienda 4:")
print(top_5_tienda4)
print("\nBottom 5 productos menos vendidos - Tienda 4:")
print(bottom_5_tienda4)
```

# 5. Envío promedio por tienda

En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.  Sabiendo que el cliente paga el envío.

```python
costo_envio_promedio_tienda1 = tienda1['Costo de envío'].mean()
costo_envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
costo_envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
costo_envio_promedio_tienda4 = tienda4['Costo de envío'].mean()

print(f'Costo de envío promedio de la Tienda 1: ${format_latam(costo_envio_promedio_tienda1)}')
print(f'Costo de envío promedio de la Tienda 2: ${format_latam(costo_envio_promedio_tienda2)}')
print(f'Costo de envío promedio de la Tienda 3: ${format_latam(costo_envio_promedio_tienda3)}')
print(f'Costo de envío promedio de la Tienda 4: ${format_latam(costo_envio_promedio_tienda4)}')
```
